```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=hp.get_data('e0103.csv')
```

Detecting peaks in the signal using the library Scipy in order to increase the accuracy

```python
import numpy as np
from scipy.signal import find_peaks,detrend
import matplotlib.pyplot as plt
peaks, _ = find_peaks(data, prominence=1)
peak_diff = np.diff(peaks)
plt.figure(figsize=(10, 6))
plt.plot(data, label='Data')
plt.plot(peaks, data[peaks], 'x', color='red', label='Peaks')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Peak Detection')
plt.legend()
plt.grid(True)
plt.show()
```

```python
breath_peaks, _ = find_peaks(peak_diff, prominence=1)
breath_peaks_diff = np.diff(breath_peaks)
plt.figure(figsize=(10, 6))
plt.plot(peak_diff)
plt.plot(breath_peaks, peak_diff[breath_peaks], 'x', color='red')
plt.title('Peak Detection')
plt.show()
```
